```python
import matplotlib.pyplot as plt
import pandas as pd
import os


def plot_loss(ax, title, df, gen_color='blue', disc_color='red'):
    """Plots the losses of given GAN loss file.

    Args:
        ax (matplotlib.axes.Axes):
           The axes of the figure subplot.
        title (str):
            The title of the subplot.
        gen_color (str):
            Color of the generator loss curve. Defaults to blue.
        disc_color (str):
            Color of the discriminator loss curve. Defaults to red.
    """
    ax.set_xlabel('Epochs')
    ax.set_ylabel('Loss')
    ax.set_title(title)
    df.generator_loss.plot(label='Generator loss', color=gen_color)
    df.discriminator_loss.plot(label='Discriminator loss', color=disc_color)
    plt.legend()


fig = plt.figure(figsize=(60, 16))
cols = 5

loss_folder = '../losses/'

ax1 = fig.add_subplot(2, cols, 1)
df1 = pd.read_csv(os.path.join(loss_folder, 'dp_cgans_loss_output_10000.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax1, 'DP-CGAN 2e-4 lr 1 critic step', df1, gen_color='orange')

ax2 = fig.add_subplot(2, cols, 2)
df2 = pd.read_csv(os.path.join(loss_folder, 'brain_lung_100_50_5000_epochs_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax2, 'Onto-DP-CGAN 2e-4 lr 5 critic steps only RD embed 100x50 100 noise fixed loss UNSEEN', df2, gen_color='green')

ax3 = fig.add_subplot(2, cols, 3)
df3 = pd.read_csv(os.path.join(loss_folder, 'leukemia_600_50_10000_epochs_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax3, 'Onto-DP-CGAN 2e-4 lr 5 critic steps only RD embed 600x50 100 noise fixed loss UNSEEN', df3, gen_color='green')

ax4 = fig.add_subplot(2, cols, 4)
df4 = pd.read_csv(os.path.join(loss_folder, 'wrong_loss_600_41_10000_epochs_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax4, 'Onto-DP-CGAN 2e-4 lr 5 critic steps only RD embed 600x41 100 noise', df4, gen_color='#42c5f5')

ax5 = fig.add_subplot(2, cols, 5)
df5 = pd.read_csv(os.path.join(loss_folder, 'wrong_loss_1080_112_10000_epochs_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax5, 'Onto-DP-CGAN 2e-4 lr 5 critic steps all embeds 1080x112 100 noise', df5, gen_color='#26b5b0')

ax6 = fig.add_subplot(2, cols, 6)
df6 = pd.read_csv(os.path.join(loss_folder, 'all_embeds_600_41_10000_epochs_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax6, 'Onto-DP-CGAN 2e-4 lr 5 critic steps all embeds 600x41 100 noise FIXED LOSS', df6, gen_color='#e618ed')

ax7 = fig.add_subplot(2, cols, 7)
df7 = pd.read_csv(os.path.join(loss_folder, 'all_embeds_600_41_5000_epochs_300_noise_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax7, 'Onto-DP-CGAN 2e-4 lr 5 critic steps all embeds 600x41 300 noise FIXED LOSS', df7, gen_color='#e618ed')

ax8 = fig.add_subplot(2, cols, 8)
df8 = pd.read_csv(os.path.join(loss_folder, 'rd_embedding_600_50_10000_epochs_loss_output.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax8, 'Onto-DP-CGAN 2e-4 lr 5 critic steps only RD 600x50 100 noise FIXED LOSS', df8, gen_color='#e618ed')

ax9 = fig.add_subplot(2, cols, 9)
df9 = pd.read_csv(os.path.join(loss_folder, 'no_embedding_600_50_10000_epochs_loss_output.txt'), sep=',', header=0).iloc[::200]
plot_loss(ax9, 'Onto-DP-CGAN 2e-4 lr 5 critic steps NO EMBED 600x50 100 noise FIXED LOSS', df9, gen_color='#e618ed')

ax10 = fig.add_subplot(2, cols, 10)
df10 = pd.read_csv(os.path.join(loss_folder, 'old_onto_cgan_loss_output_10000.txt'), sep=',', header=0).iloc[::100]
plot_loss(ax10, 'OLD Onto-DP-CGAN 2e-4 lr 1 critic step only rd embed', df10, gen_color='blue')
```
